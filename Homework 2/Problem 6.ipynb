```python
# 2
N = 50
mu = 3.8
x = np.zeros(N)
x[0] = 0.8
for i in range(N - 1):
    x[i+1] = mu * x[i] * (1 - x[i])
plt.scatter(np.linspace(0, N, N), x)
plt.plot(np.linspace(0, N, N), x)
plt.grid()
plt.show()
```

```python
def logistic_equation(mu, x0, N):
    x = np.zeros(N)
    x[0] = x0
    for i in range(N - 1):
        x[i+1] = mu * x[i] * (1 - x[i])
    return x
```

```python
N = 100
mu = np.linspace(0, 4.0, N)
mus = []
values = []
for i in range(len(mu)):
    x = logistic_equation(mu[i], 0.8, N)
    mus.extend([mu[i]] * (N - 10))
    values.extend(x[10:])
plt.scatter(mus, values, c = 'k', marker= '.', s = 1)
plt.show()
```

```python
def random_number(order):
    x = logistic_equation(4, 0.8, 100)
    r = rand.choice(x) * order
    return r
```

```python
print(random_number(10))
```
