```python
import numpy as np 
import matplotlib.pyplot as plt
```

# **Using Linear Interpolation**

```python
x = np.linspace(-10, 10, 10)
noise = np.random.randint(0, 10)
y = x ** 2 + noise
plt.scatter(x, y)
for i in range(len(x) - 1):
    x1 = x[i]
    x2 = x[i+1]
    y1 = x1 ** 2 + noise
    y2 = x2 ** 2 + noise
    a = (y2 - y1)/(x2 - x1)
    b = y1 - a * x1
    y = a * x + b 
    plt.plot(x, y)
plt.ylim(-10, 115)
plt.show()
```

# **Using Lagrange Interpolation**

## 2nd Order Lagrange Interpolation

```python
x = np.linspace(-10, 10, 100)
noise = np.random.randint(0, 10)
y = x ** 2 + noise
plt.plot(x,y, color = 'k')
x = np.linspace(-10, 10, 10)
noise = np.random.randint(0, 10)
y = x ** 2 + noise
plt.scatter(x, y)
for i in range(len(x) - 2):
    x1 = x[i]
    x2 = x[i+1]
    x3 = x[i+2]
    y1 = x1 ** 2 + noise
    y2 = x2 ** 2 + noise
    y3 = x3 ** 2 + noise
    L1 = ((x - x2)/(x1 - x2)) * ((x - x3)/(x1 - x3))
    L2 = ((x - x1)/(x2 - x1)) * ((x - x3)/(x2 - x3))
    L3 = ((x - x1)/(x3 - x1)) * ((x - x2)/(x3 - x2))
    y = L1 * y1 + L2 * y2 + L3 * y3
    plt.plot(x, y)
plt.ylim(-10, 115)
plt.show()
```
