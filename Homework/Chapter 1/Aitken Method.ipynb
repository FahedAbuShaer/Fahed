```python
import numpy as np
import matplotlib.pyplot as plt
```

# **The Aitken Method for Interpolation** 

```python
def Aitkin (x, P, first, last):
    if first == last:
        return P[first]
    elif first != last:
        return ((x - x[last])/(x[first] - x[last])) * Aitkin(x, P, first, last - 1) + ((x - x[first])/(x[last] - x[first])) * Aitkin(x, P, first + 1, last)
```

```python
x = np.linspace(-20, 20, 50)
y = 1/(1 + x ** 2)
plt.plot(x, y, color = 'k')
x = np.linspace(-20, 20, 10)
y = 1/(1 + x ** 2)
plt.scatter(x, y)
P = Aitkin(x, y, 0, len(x) - 1)
plt.plot(x, P)
plt.show()
```
