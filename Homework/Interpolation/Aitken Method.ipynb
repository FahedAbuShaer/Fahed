```python
import numpy as np
import matplotlib.pyplot as plt
```

# **The Aitken Method for Interpolation** 

```python
def Aitkin (x, P, first, last):
    if first == last:
        return P[first]
    elif first != last:
        return ((x - x[last])/(x[first] - x[last])) * Aitkin(x, P, first, last - 1) + ((x - x[first])/(x[last] - x[first])) * Aitkin(x, P, first + 1, last)
```

```python
x = np.linspace(-50, 50, 200)
y = 1/(1 + x ** 2)
plt.plot(x, y, color = 'k')
x = np.linspace(-50, 50, 20)
y = 1/(1 + x ** 2)
plt.scatter(x, y)
for i in range(len(x) - 1):
    xs = np.linspace(x[i], x[i+1], 10)
    ys = 1/(1 + xs ** 2)
    Ps = Aitkin(xs, ys, 0, len(xs) - 1)
    plt.plot(xs, Ps, color = 'r')
plt.show()
```
