```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def trap_integral(x, y, h):
    I = np.sum((y[i] + y[i+1]) * h / 2)
    return I
```

```python
l = 100
I = []
h = []
for i in range(1, l + 1, 10):
    n = 10 * i
    x = np.linspace(-10, 10, n)
    y = x**2
    d = x[1] - x[0]
    Int = trap_integral(x, y, d)
    h.append(d)
    I.append(Int)
    error = np.abs(666.67 - Int)
    print("h = ", d, " n = ", n, " I = ", Int, " Error = ", error)
plt.plot(h, I)
plt.scatter(h, I)
plt.axis([h[0], h[-1], I[-1] - 2, I[0] + 3])
plt.ylabel("Numerical Integral")
plt.xlabel("Distance between two points")
plt.show()
```
