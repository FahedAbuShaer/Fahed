```python
import numpy as np
import matplotlib.pyplot as plt 
import networkx as nx
```

```python
l = [(1,3), (2,4), (4,6), (1,6), (5,4), (5,1), (3,1), (4,6), (5, 6)]
g = nx.Graph(l)
nx.draw_spring(g, with_labels = True)
plt.show()
c = nx.DiGraph(l)
nx.draw_planar(c, with_labels = True)
plt.show()
```
