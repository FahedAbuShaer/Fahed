# Exercise 2: The Collatz conjecture

The Conjecture is the following :
Given a number n $\in \mathbb{N}^+$ if it is even divide it by 2 and if it is odd multiply it by 3 and add 1. 
According to the conjecture this will always converge to 1. 

Write a code that calculates the steps of the Collatz Conjecture till it reches 1 and plot them. 



```python
def conjecture(n):
    list = [n]
    while n != 1:
        if n % 2 == 0:
            n = n/2
            list.append(n)
        else:
            n = 3 * n + 1
            list.append(n)
    plt.plot(list)
    plt.show()

conjecture(5)
```

# Exercise 3: Benford's Law. 

Benfords law describes the relative frequency distribution for leading digits of numbers in datasets. Leading digits with smaller values occur more frequently than larger values. This law states that approximately 30% of numbers start with a 1 while less than 5% start with a 9. According to this law, leading 1s appear 6.5 times as often as leading 9s! Benfords law is also known as the First Digit Law.

We will be trying this using the world population for each contry in 'word_population.csv'.

# Linear Algebra

$$
x + 5y + 5z = 18\\
8x + 14y -6z = 46\\
4x + 6y + 9z = 33\\
$$

Solve the system


```python
A = np.array([[1,5,5], [8,14,-6],[4,6,9]])
b = np.array([18, 46, 33])
print(np.linalg.solve(A, b))
```

$$
x + 2y = 4\\
x + 2y = 5
$$
Solve the system with the best solution possible

```python
A = np.array([[1, 2], [1, 2]])
B = np.array([4, 5])
x = np.linalg.lstsq(A, B)
print(x)
```
