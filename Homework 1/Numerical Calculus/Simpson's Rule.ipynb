```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def least_square_approximation(x, y, n):
    V = np.empty([len(x), n + 1])
    for i in range(n + 1):
        V[:, i] = x ** i
    Vt = np.transpose(V)
    A = Vt @ V
    Y = Vt @ y
    b = np.linalg.solve(A, Y)
    return b
```

```python
def fitpolynomial(x, b, n):
    Ps = np.zeros(len(x))
    for j in range(len(x)):
        p = 0
        for i in range(n + 1):
            p += b[i] * x[j] ** i
        Ps[j] = p
    return Ps
```

```python
x = np.linspace(-10, 10, 10)
y = x**2
plt.scatter(x, y)
c = least_square_approximation(x, y, 2)
for i in range(len(x) - 1):
        xs = np.linspace(x[i], x[i+1], 50)
        p = fitpolynomial(xs, c, 2)
        plt.plot(xs,p, color = "violet")
plt.show()
```

```python
x =np.linspace(-10, 10, 10000000)
y = x**2
h = x[1] - x[0]
print(h)
b = least_square_approximation(x, y, 2)
l = fitpolynomial([x[0] - h], b, 2)
I = ((4/3) * h * y[0]) + ((h/3) * l) + ((h/3) * y[1])
for i in range(1, len(y) - 1, 2):
    I += ((4/3) * h * y[i]) + ((h/3) * y[i-1]) + ((h/3) * y[i+1])
k = fitpolynomial([x[len(x) - 1] + h], b, 2)
I += ((4/3) * h * y[len(y) - 1]) + ((h/3)* y[len(y) - 2]) + ((h/3) * k)
print(I)
```

```python
def SimpsonsRule(x, y, h, n):
    b = least_square_approximation(x, y, n)
    l = fitpolynomial([x[0] - h], b, n)
    I = ((4/3) * h * y[0]) + ((h/3) * l) + ((h/3) * y[1])
    for i in range(1, len(y) - 1, 2):
        I += ((4/3) * h * y[i]) + ((h/3) * y[i-1]) + ((h/3) * y[i+1])
    k = fitpolynomial([x[len(x) - 1] + h], b, n)
    I += ((4/3) * h * y[len(y) - 1]) + ((h/3)* y[len(y) - 2]) + ((h/3) * k)
    return I
```

```python
for i in range(1, 20):
    x = np.linspace(-10, 10, 10*i)
    y=x**2
    h = x[1] - x[0]
    print("h = ", h)
    I = SimpsonsRule(x, y, h)
    print("I = ", I)
```

```python
l = 10000
I = []
h = []

x1 = np.linspace(-10,10,10)
y1 = x1**2
d1 = x1[1] - x1[0]
Inte = SimpsonsRule(x1, y1, d1, 2)
error = np.abs(666.67 - Inte[0])
print("h = ", d1, " I = ", Inte[0], " Error = ", error)

for i in range(100, l + 1, 500):
    n = 10 * i
    x = np.linspace(-10, 10, n)
    y = x**2
    d = x[1] - x[0]
    Int = SimpsonsRule(x, y, d, 2)
    h.append(d)
    I.append(Int[0])
    error = np.abs(666.67 - Int[0])
    print("h = ", d, " I = ", Int[0], " Error = ", error)
plt.plot(h, I)
plt.scatter(h, I)
plt.axis([h[0], h[-1], I[-1] - 2, I[0] + 3])
plt.ylabel("Numerical Integral")
plt.xlabel("Distance between two points")
plt.show()
print(h, "\n", I)
```
