```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Vandermond(x, y, n):
    V = np.empty([len(x), n + 1])
    for i in range(len(x) - 1):
        for j in range(n + 1):
            V[i, j] = x[i] ** (n - j)
    return V
```

```python
def LinearRegression(V, y):
    Vt = np.transpose(V)
    A = Vt @ V
    Y = Vt @ y
    b = np.linalg.solve(A, Y)
    return b
```

```python
x = np.linspace(-10, 10, 20)
y = np.random.uniform(0, 10, 20) * x + np.random.uniform(0, 10, 20)
plt.scatter(x, y)
n = 2
V = Vandermond(x, y, n)
b = LinearRegression(V, y)
p = b[0] + b[1] * x
plt.plot(x, p)
plt.show()
```
